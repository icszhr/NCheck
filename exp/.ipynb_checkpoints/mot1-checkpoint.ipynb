```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
import matplotlib as mpl

# 确保 Times New Roman 字体可用
font_path = '/usr/share/fonts/truetype/msttcorefonts/Times_New_Roman.ttf'  # 替换为实际路径
try:
    fm.fontManager.addfont(font_path)
    mpl.rcParams['font.family'] = 'Times New Roman'
except FileNotFoundError:
    print("Times New Roman font not found, using default serif font.")
    mpl.rcParams['font.family'] = 'serif'
    mpl.rcParams['font.serif'] = 'DejaVu Serif'

# 数据
models = ["ResNet50", "GPT-2", "Bert", "RoBERTa", "T5", "DistilBERT"]
forward_time = [0.04, 0.23, 0.20, 0.49, 0.14, 0.10]
backward_time = [0.06, 0.35, 0.39, 0.66, 0.17, 0.21]
weight_time = [0.02, 0.07, 0.11, 0.25, 0.14, 0.07]
snapshot_time = [0.08, 0.54, 0.41, 1.48, 0.20, 0.33]
persist_time = [1.14, 2.67, 3.25, 5.59, 2.64, 1.91]

# 设置字体和样式
fontSize = 14 / 1.2

mpl.rcParams.update({
    "font.size": fontSize,
    "axes.labelsize": fontSize * 1.5,
    "xtick.labelsize": fontSize * 1.3,
    "ytick.labelsize": fontSize * 1.3,
    "legend.fontsize": fontSize * 1.3,
    "axes.titlesize": fontSize * 1.5,
    "axes.linewidth": 0.75,
    "xtick.major.width": 0.75,
    "ytick.major.width": 0.75,
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "lines.linewidth": 1,
    "lines.markersize": 3,
    "figure.figsize": (4, 2.5),
})

# 增强辨识度的莫兰迪配色
morandi_colors = ['#fffaf4', '#B5B5B5', '#CCCCCC', '#F8CECC', '#E1D5E7']

# 绘图
fig, ax = plt.subplots(figsize=(8, 5), linewidth=0.75)

bar_width = 0.35
index = np.arange(len(models))

# 绘制堆叠柱状图
p1 = plt.bar(models, forward_time, bar_width, label='Forward Propagation', color='w', edgecolor='black')
p2 = plt.bar(models, backward_time, bar_width, bottom=forward_time, label='Backward Propagation', color='w', edgecolor='black', hatch = '//')
p3 = plt.bar(models, weight_time, bar_width, bottom=np.array(forward_time) + np.array(backward_time), label='Weight Update', color=morandi_colors[2], edgecolor='black')
p4 = plt.bar(models, snapshot_time, bar_width, bottom=np.array(forward_time) + np.array(backward_time) + np.array(weight_time), label='Snapshot', color=morandi_colors[3], edgecolor='black')
p5 = plt.bar(models, persist_time, bar_width, bottom=np.array(forward_time) + np.array(backward_time) + np.array(weight_time) + np.array(snapshot_time), label='Persist', color=morandi_colors[4], edgecolor='black')

# 添加标签和标题
ax.set_ylabel('Time (seconds)', fontsize=fontSize * 1.5, weight='bold')
ax.set_xticks(index)
ax.set_xticklabels(models, fontsize=fontSize * 1.3)
ax.legend(fontsize=fontSize * 1, loc='upper left', bbox_to_anchor=(0.1, 1), ncol=1, frameon=False)

# 设置刻度字体大小
ax.tick_params(axis='both', which='major', labelsize=fontSize * 1.3)

# 显示图表
plt.show()
```
