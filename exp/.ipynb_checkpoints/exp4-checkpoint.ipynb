```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
import matplotlib as mpl

# 确保 Times New Roman 字体可用
font_path = '/usr/share/fonts/truetype/msttcorefonts/Times_New_Roman.ttf'  # 替换为实际路径
try:
    fm.fontManager.addfont(font_path)
    mpl.rcParams['font.family'] = 'Times New Roman'
except FileNotFoundError:
    print("Times New Roman font not found, using default serif font。")
    mpl.rcParams['font.family'] = 'serif'
    mpl.rcParams['font.serif'] = 'DejaVu Serif'

# 设置数据
x_labels = ['Checkpoint only', 'Checkpoint+Swap', 'NCheck']

# 计算每根柱子的大小
group1 = [3470]
group2 = [78.9 + 2441 + 520/4 + 42/4, 78.9 + 2441  + 0.75*520 +0.75*42]
group3 = [2441 + 562/6, 0.25*520 + 79 + 2441, 79 + 0.75*562 + 79 + 2441+ 2000]

# 设置柱子的宽度
bar_width = 0.4

# 设置x轴位置，每组之间留有间隔
x = np.array([0, 1.5, 3])  # 增加间距确保间隔分布

# 设置字体和样式
fontSize = 14 / 1.2

mpl.rcParams.update({
    "font.size": fontSize,
    "axes.labelsize": fontSize * 1.5,
    "xtick.labelsize": fontSize * 1.3,
    "ytick.labelsize": fontSize * 1.3,
    "legend.fontsize": fontSize * 1.3,
    "axes.titlesize": fontSize * 1.5,
    "axes.linewidth": 0.75,
    "xtick.major.width": 0.75,
    "ytick.major.width": 0.75,
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "lines.linewidth": 1,
    "lines.markersize": 3,
    "figure.figsize": (12, 4),  # 将图片高度缩小一半
})

# 设置颜色
colors = ['#B5B5B5', '#D3BFBF', '#8696a7']

# 创建绘图
fig, ax = plt.subplots(figsize=(5, 4))

# 绘制每组数据，按照颜色和顺序要求
bars1 = ax.bar(x[0], group1[0], color=colors[0], width=bar_width, edgecolor='grey', label='1st bar')
bars2 = ax.bar(x[1] - bar_width/2, group2[0], color=colors[0], width=bar_width, edgecolor='grey')
bars3 = ax.bar(x[1] + bar_width/2, group2[1], color=colors[1], width=bar_width, edgecolor='grey')
bars4 = ax.bar(x[2] - bar_width, group3[0], color=colors[0], width=bar_width, edgecolor='grey')
bars5 = ax.bar(x[2], group3[1], color=colors[1], width=bar_width, edgecolor='grey')
bars6 = ax.bar(x[2] + bar_width, group3[2], color=colors[2], width=bar_width, edgecolor='grey')

# 添加x轴标签和刻度
ax.set_ylabel('Peak Memory Usage (MB)', fontsize=fontSize * 1.5, weight='bold')
ax.set_xticks(x)
ax.set_xticklabels(x_labels)

# 设置图例
ax.legend([bars1, bars3, bars6], ['GPU Mem', 'DRAM', 'NVM'], fontsize=fontSize * 1.3, loc='upper left', frameon=False)

# 设置刻度字体大小
ax.tick_params(axis='both', which='major', labelsize=fontSize * 1.5)

# 调整x轴刻度位置使其对齐到每组柱子的中间
ax.set_xticks([0, 1.5, 3])
ax.set_xticklabels(['CheckFreq', 'GEMINI', 'NCheck'])

# 显示图表
plt.tight_layout()
plt.savefig('./exp4.pdf', bbox_inches='tight')
plt.show()
```
