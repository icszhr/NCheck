```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
import matplotlib as mpl

#  Times New Roman 
font_path = '/usr/share/fonts/truetype/msttcorefonts/Times_New_Roman.ttf'  # 
try:
    fm.fontManager.addfont(font_path)
    mpl.rcParams['font.family'] = 'Times New Roman'
except FileNotFoundError:
    print("Times New Roman font not found, using default serif font")
    mpl.rcParams['font.family'] = 'serif'
    mpl.rcParams['font.serif'] = 'DejaVu Serif'

# 
base_data = [12.75, 24.23, 24.64, 25.46, 24.8, 13]

# 
top_data = [
    [0.3, 0.58 * 5, 0.52 * 5, 0.37 * 5/2],  # 
    [0.24 * 5, 1.38 * 5, 1.18 * 5, 1.01 * 5/2],  # 
    [0.22 * 5, 1.29 * 5, 1.04 * 5, 0.99 * 5/2],  # 
    [0.27 * 5, 1.54 * 5, 1.33 * 5, 1.11 * 5/2],  # 
    [0.18 * 5, 1.36 * 5, 1.26 * 5, 1.02 * 5/2],  # 
    [0.29, 0.79 * 5, 0.7 * 5, 0.46 * 5/2]   # 
]

# 
models = ['ResNet50', 'GPT2', 'BERT', 'RoBERTa', 'T5', 'DistilBERT']

# 
base_times = np.array(base_data) * 10
top_times = np.array(top_data) * 10 / 2  # 10

# 
fontSize = 14 / 1.2

mpl.rcParams.update({
    "font.size": fontSize,
    "axes.labelsize": fontSize * 1.5,
    "xtick.labelsize": fontSize * 1.3,
    "ytick.labelsize": fontSize * 1.3,
    "legend.fontsize": fontSize * 1.3,
    "axes.titlesize": fontSize * 1.5,
    "axes.linewidth": 0.75,
    "xtick.major.width": 0.75,
    "ytick.major.width": 0.75,
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "lines.linewidth": 1,
    "lines.markersize": 3,
    "figure.figsize": (12, 4),  # 
})

# hatch
colors = ['#000000', '#FFFFFF', '#555555', '#777777']
hatches = ['', '', '////', 'xx']

# 
fig, ax = plt.subplots(figsize=(8, 3.45))

# 
x = np.arange(len(models))
width = 0.2
ax.set_ylim(0, 450)  # 

# 
for i in range(len(models)):
    ax.bar(x[i] + np.arange(4) * width, [base_times[i]] * 4, width, color='#CCCCCC', edgecolor='black')

# 
bars = []
for j in range(4):
    bar = ax.bar(x + j * width, top_times[:, j], width, bottom=base_times, color=colors[j], edgecolor='black', hatch=hatches[j])
    bars.append(bar)

# 
for i in range(len(models)):
    for j in range(4):
        total_height = base_times[i] + top_times[i, j]
        percentage = top_times[i, j] / total_height * 100
        if j % 2 == 0:
            y_offset = 10  # 
        else:
            y_offset = 25  # 

        ax.text(x[i] + j * width, total_height + y_offset, f'{percentage:.1f}', ha='center', va='bottom', fontsize=fontSize * 0.9)
        ax.plot([x[i] + j * width, x[i] + j * width], [total_height, total_height + y_offset + 5], color='black')

# 
ax.legend([bars[0], bars[1], bars[2], bars[3]], ['NCheck', 'Origin-TF', 'CheckFreq', 'LightCheck'], fontsize=fontSize* 1.3, loc='upper left', frameon=False, ncol=4, columnspacing=0.5, bbox_to_anchor=(0.015, 1.05))

# 
ax.set_ylabel('Time (seconds)', fontsize=fontSize * 1.5, weight='bold')
ax.set_xticks(x + width * 1.5)
ax.set_xticklabels(models)

# 
ax.tick_params(axis='both', which='major', labelsize=fontSize * 1.5)

# 
plt.tight_layout()
plt.savefig('./exp1.pdf', bbox_inches='tight')
plt.show()
```
