```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
import matplotlib as mpl
import matplotlib.patches as mpatches

# 确保 Times New Roman 字体可用
font_path = '/usr/share/fonts/truetype/msttcorefonts/Times_New_Roman.ttf'  # 替换为实际路径
try:
    fm.fontManager.addfont(font_path)
    mpl.rcParams['font.family'] = 'Times New Roman'
except FileNotFoundError:
    print("Times New Roman font not found, using default serif font。")
    mpl.rcParams['font.family'] = 'serif'
    mpl.rcParams['font.serif'] = 'DejaVu Serif'

# 设置数据
models = ["ResNet50", "GPT-2", "BERT", "RoBERTa", "T5", "DistilBERT"]

# 随机生成0到2之间的数据
total_times = np.array([
    [1, 0.005/0.12, 0.17/0.12, 0.068/0.12],
    [1, 0.8, 0.26/0.26, 0.047],
    [1, 0.6, 0.077/0.3, 0.1],
    [1, 0.9, 0.089/0.28, 0.047/0.28],
    [1, 0.5, 0.17/0.25, 0.1],
    [1, 1.0, 0.04/0.14, 0.4]
])
#exe: 0.12, 0.27, 0.30, 0.28, 0.25, 0.14
# 设置字体和样式
fontSize = 14 / 1.2

mpl.rcParams.update({
    "font.size": fontSize,
    "axes.labelsize": fontSize * 1.5,
    "xtick.labelsize": fontSize * 1.2,
    "ytick.labelsize": fontSize * 1.2,
    "legend.fontsize": fontSize * 1.3,
    "axes.titlesize": fontSize * 1.5,
    "axes.linewidth": 0.75,
    "xtick.major.width": 0.75,
    "ytick.major.width": 0.75,
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "lines.linewidth": 1,
    "lines.markersize": 3,
    "figure.figsize": (12, 4),  # 将图片高度缩小一半
})

# 增强辨识度的莫兰迪配色
morandi_colors = ['#B5B5B5', '#D3BFBF', '#8696a7', '#aab8ab']

# 设置柱状图
fig, ax = plt.subplots(figsize=(8, 4))

# 设置位置和宽度
x = np.arange(len(models))
width = 0.2
ax.set_ylim(0, 2)  # 设置y轴的上限为2.5

# 绘制柱状图
bars = []
labels = ['Base(No Checkpoint)', 'Parallelism Blocking', 'Metadata Update', 'Log GC']
for i in range(total_times.shape[1]):
    bar = ax.bar(x + i*width, total_times[:, i], width, color=morandi_colors[i], edgecolor='black', label=f'{labels[i]}')
    bars.append(bar)

# 仅选择需要的句柄
handles, labels = ax.get_legend_handles_labels()
selected_handles = [handles[i] for i in range(len(bars))]

# 设置图例
ax.legend(selected_handles, labels, fontsize=fontSize * 1.2, loc='upper center', frameon=False, ncol=2)

# 添加标签和标题
ax.set_ylabel('Normalized Time', fontsize=fontSize * 1.5, weight='bold')
ax.set_xticks(x + width*1.5)
ax.set_xticklabels(models)

# 设置刻度字体大小
ax.tick_params(axis='both', which='major', labelsize=fontSize * 1.5)

# 显示图表
plt.tight_layout()
plt.savefig('./exp9.pdf', bbox_inches='tight')
plt.show()
```
