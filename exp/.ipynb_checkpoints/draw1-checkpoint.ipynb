```python
import json
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
import matplotlib as mpl

# 确保 Times New Roman 字体可用
font_path = '/usr/share/fonts/truetype/msttcorefonts/Times_New_Roman.ttf'  # 替换为实际路径
try:
    fm.fontManager.addfont(font_path)
    mpl.rcParams['font.family'] = 'Times New Roman'
except FileNotFoundError:
    print("Times New Roman font not found, using default serif font.")
    mpl.rcParams['font.family'] = 'serif'
    mpl.rcParams['font.serif'] = 'DejaVu Serif'

# 读取结果文件
with open('checkpoint_results.json', 'r') as f:
    results = json.load(f)

# 模型名称
models = ["GPT2", "Bert", "Roberta", "DistilBert"]

# 数据
ssd_save_times = [results[model]['ssd_save_time'] for model in results.keys()]
ssd_serialize_times = [results[model]['ssd_serialize_time'] for model in results.keys()]
nvm_save_times = [results[model]['nvm_save_time'] for model in results.keys()]
ssd_restore_times = [results[model]['ssd_restore_time'] for model in results.keys()]
ssd_deserialize_times = [results[model]['ssd_deserialize_time'] for model in results.keys()]
nvm_restore_times = [results[model]['nvm_restore_time'] for model in results.keys()]

# 设置字体和样式
fontSize = 14 / 1.2

mpl.rcParams.update({
    "font.size": fontSize,
    "axes.labelsize": fontSize * 1.5,
    "xtick.labelsize": fontSize * 1.3,
    "ytick.labelsize": fontSize * 1.3,
    "legend.fontsize": fontSize * 1.3,
    "axes.titlesize": fontSize * 1.5,
    "axes.linewidth": 0.75,
    "xtick.major.width": 0.75,
    "ytick.major.width": 0.75,
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "lines.linewidth": 1,
    "lines.markersize": 3,
    "figure.figsize": (4, 2.5),
})

# 增强辨识度的莫兰迪配色
morandi_colors = ['#fffaf4', '#B5B5B5', '#D3BFBF', '#8696a7', '#F5EDED', '#B19B9B', '#A58D8D', '#937575', '#d8caaf']

# 绘图
fig, ax = plt.subplots(figsize=(8, 3.5), linewidth=0.75)

bar_width = 0.2
index = np.arange(len(models))

# SSD 保存时间
bar1 = ax.bar(index, ssd_save_times, bar_width, label='SSD Save', color=morandi_colors[0], edgecolor='black', hatch='//')
# SSD 序列化时间
bar1_serialize = ax.bar(index, ssd_serialize_times, bar_width, label='SSD Serialize', color=morandi_colors[1], edgecolor='black', hatch='\\\\', bottom=ssd_save_times)
# NVM 保存时间
bar2 = ax.bar(index + bar_width, nvm_save_times, bar_width, label='NVM Save', color=morandi_colors[2], edgecolor='black', hatch='--')

# SSD 恢复时间
bar3 = ax.bar(index + 2 * bar_width, ssd_restore_times, bar_width, label='SSD Recovery', color=morandi_colors[3], edgecolor='black', hatch='..')
# SSD 反序列化时间
bar3_deserialize = ax.bar(index + 2 * bar_width, ssd_deserialize_times, bar_width, label='SSD Deserialize', color=morandi_colors[4], edgecolor='black', hatch='++', bottom=ssd_restore_times)
# NVM 恢复时间
bar4 = ax.bar(index + 3 * bar_width, nvm_restore_times, bar_width, label='NVM Recovery', color=morandi_colors[5], edgecolor='black', hatch='xx')

# 添加标签和标题
#ax.set_xlabel('Model', fontsize=fontSize * 1.5, weight='bold')
ax.set_ylabel('Time (seconds)', fontsize=fontSize * 1.5, weight='bold')
#ax.set_title('Checkpoint Performance for Different Models', fontsize=fontSize * 1.5, weight='bold')
ax.set_xticks(index + 1.5 * bar_width)
ax.set_xticklabels(models, fontsize=fontSize * 1.3)
ax.legend(fontsize=fontSize * 1, loc='upper center', bbox_to_anchor=(0.55, 1.03), ncol=1, frameon=False)

# 设置刻度字体大小
ax.tick_params(axis='both', which='major', labelsize=fontSize * 1.3)

# 显示图表
plt.savefig('./checkpoint_performance.pdf', bbox_inches='tight')
plt.show()
```
