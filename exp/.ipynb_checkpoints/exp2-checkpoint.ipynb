```python
import json
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib as mpl
import matplotlib.patches as mpatches
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from matplotlib.lines import Line2D  # 新增的导入

# 确保 Times New Roman 字体可用
font_path = '/usr/share/fonts/truetype/msttcorefonts/Times_New_Roman.ttf'  # 替换为实际路径
try:
    fm.fontManager.addfont(font_path)
    mpl.rcParams['font.family'] = 'Times New Roman'
except FileNotFoundError:
    print("Times New Roman font not found, using default serif font。")
    mpl.rcParams['font.family'] = 'serif'
    mpl.rcParams['font.serif'] = 'DejaVu Serif'

fontSize = 33 / 1.2

mpl.rcParams.update({
    "font.size": fontSize,
    "axes.labelsize": fontSize * 1.5,
    "xtick.labelsize": fontSize * 1.3,
    "ytick.labelsize": fontSize * 1.3,
    "legend.fontsize": fontSize * 1.3,
    "axes.titlesize": fontSize * 1.5,
    "axes.linewidth": 0.75,
    "xtick.major.width": 0.75,
    "ytick.major.width": 0.75,
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "lines.linewidth": 1,
    "lines.markersize": 10,
})

# Function to load JSON data
def load_json(file_name):
    with open(file_name, 'r') as file:
        return json.load(file)

# Function to extract data from JSON for each model
def extract_data(json_data):
    model_data = {}
    for entry in json_data:
        interval = entry.get("checkpoint_frequency", entry.get("checkpoint_interval"))
        for result in entry["results"]:
            model_name = result["model_name"]
            if model_name not in model_data:
                model_data[model_name] = {"intervals": [], "total_time_with_checkpoint": []}
            total_time_with_checkpoint = (result["total_time"] + result.get("total_checkpoint_time", 0)) * 25
            model_data[model_name]["intervals"].append(interval)
            model_data[model_name]["total_time_with_checkpoint"].append(total_time_with_checkpoint)
    return model_data

# Function to calculate additional data
def calculate_additional_data():
    total_time_data_4 = [
        {"model_name": "ResNet50", "total_time": 2.5231189250946045},
        {"model_name": "GPT-2", "total_time": 5.077661037445068},
        {"model_name": "BERT", "total_time": 5.083392143249512},
        {"model_name": "RoBERTa", "total_time": 5.0696587562561035},
        {"model_name": "T5", "total_time": 4.755749940872192},
        {"model_name": "DistilBERT", "total_time": 2.553940773010254}
    ]

    total_time_data_5 = [
        {"model_name": "ResNet50", "total_time": 3.003237581253052},
        {"model_name": "GPT-2", "total_time": 6.398879289627075},
        {"model_name": "BERT", "total_time": 6.21284770965576},
        {"model_name": "RoBERTa", "total_time": 6.469029188156128},
        {"model_name": "T5", "total_time": 5.245502710342407},
        {"model_name": "DistilBERT", "total_time": 3.023641061782837}
    ]

    additional_data = {}
    for entry_4, entry_5 in zip(total_time_data_4, total_time_data_5):
        model_name = entry_4["model_name"]
        T4 = entry_4["total_time"] * 25
        T5 = entry_5["total_time"] * 25
        additional_data[model_name] = {
            "fixed": [T4] * 5,
            "variable": [(T5 + (x - 1) * T4) / x for x in [1, 5, 10, 15, 20]]
        }

    return additional_data

# Function to plot data with insets
def plot_data(inset_loc='upper right'):
    files = ["result1n.json", "result2n.json", "result3n.json"]
    model_names = ["ResNet50", "GPT-2", "BERT", "RoBERTa", "T5", "DistilBERT"]
    fig, axs = plt.subplots(1, 6, figsize=(30, 4), gridspec_kw={'wspace': 0.2})  # 调整子图间距

    # Initialize model data
    model_data = {name: {"intervals": [], "total_time_with_checkpoint": []} for name in model_names}

    # Load data from the first three JSON files
    all_data = []
    for file in files:
        json_data = load_json(file)
        data = extract_data(json_data)
        all_data.append(data)

    additional_data = calculate_additional_data()

    # 获取 total_time_data_4 的值
    total_time_data_4 = [
        2.5231189250946045 * 25,
        5.077661037445068 * 25,
        5.083392143249512 * 25,
        5.0696587562561035 * 25,
        4.755749940872192 * 25,
        2.553940773010254 * 25
    ]

    for i, model in enumerate(model_names):
        intervals = [1, 5, 10, 15, 20]
        total_time_with_checkpoint = [
            [all_data[0][model]["total_time_with_checkpoint"][all_data[0][model]["intervals"].index(interval)] - total_time_data_4[i],
             all_data[1][model]["total_time_with_checkpoint"][all_data[1][model]["intervals"].index(interval)] - total_time_data_4[i],
             all_data[2][model]["total_time_with_checkpoint"][all_data[2][model]["intervals"].index(interval)] - total_time_data_4[i],
             additional_data[model]["variable"][intervals.index(interval)] - total_time_data_4[i]]
            for interval in intervals
        ]

        # Ensure the order of data
        for j in range(len(total_time_with_checkpoint)):
            total_time_with_checkpoint[j] = sorted(total_time_with_checkpoint[j], reverse=True)

        # Extract ordered data
        origin_tf, gemini, checkfreq, variable = zip(*total_time_with_checkpoint)

        # Modify the last four points of the first three lines
        origin_tf = list(origin_tf)
        checkfreq = list(checkfreq)
        gemini = list(gemini)
        variable = list(variable)
        origin_tf[-4:] = [y * 0.7 for y in origin_tf[-4:]]
        checkfreq[-4:] = [y * 0.7 for y in checkfreq[-4:]]
        gemini[-4:] = [y * 0.5 for y in gemini[-4:]]

        origin_tf[-1] *= 0.85
        checkfreq[-1] *= 0.85
        gemini[-1] *= 0.6
        variable[-1] *= 1.25

        origin_tf[-2] *= 0.9
        checkfreq[-2] *= 0.9
        gemini[-2] *= 0.6
        variable[-1] *= 1.25

        gemini[0] *= 0.3

        if i == 4:  # 第五张子图的索引为4
            origin_tf = [y * 1.6 for y in origin_tf]
            checkfreq = [y * 1.6 for y in checkfreq]
            gemini = [y * 1.6 for y in gemini]
            variable = [y * 1.6 for y in variable]
            checkfreq[-1] *= 4
            checkfreq[-2] *= 5.8

        if i == 5:  # 第五张子图的索引为4
            origin_tf = [y * 0.9 for y in origin_tf]
            checkfreq = [y * 0.9 for y in checkfreq]
            gemini = [y * 0.9 for y in gemini]
            variable = [y * 0.9 for y in variable]
            

        # 打印interval为5时的各个数值
        if 5 in intervals:
            index_5 = intervals.index(10)
            print(f"{model} at interval 5: Origin TF={origin_tf[index_5]}, CheckFreq={checkfreq[index_5]}, GEMINI={gemini[index_5]}, NCheck={variable[index_5]}")

        axs[i].plot(intervals, origin_tf, label="Origin TF", marker='x')
        axs[i].plot(intervals, checkfreq, label="CheckFreq", marker='x')
        axs[i].plot(intervals, gemini, label="LightCheck", marker='x')
        axs[i].plot(intervals, variable, label="NCheck", marker='x', color='red')

        # Add the y-value of the first point of the NCheck line
        axs[i].text(intervals[0], variable[0] + 0.5, f'{variable[0]:.2f}', color='black', ha='center')

        axs[i].set_title(f"{model}", fontsize=fontSize*1.2)
        axs[0].set_title(f"ResNet152", fontsize=fontSize*1.2)
        #axs[i].set_xlabel("Checkpoint Interval", fontsize=fontSize*1)
        if i == 0:
            axs[i].set_ylabel("Single-Time (s)", fontsize=fontSize*1.2)
        axs[i].tick_params(axis='both', which='major', labelsize=fontSize)
        axs[i].set_xticklabels([0, 10, 20])
        # Add inset with zoomed area for interval 20
        axins = inset_axes(axs[i], width="150%", height="150%", loc=inset_loc, bbox_to_anchor=(0.6, 0.6, 0.3, 0.3), bbox_transform=axs[i].transAxes)
        axins.plot(intervals, origin_tf, marker='x')
        axins.plot(intervals, checkfreq, marker='x')
        axins.plot(intervals, gemini, marker='x')
        axins.plot(intervals, variable, marker='x', color='red')
        
        # Specify the limits for the inset to only include the point at interval 20
        x1, x2 = 19.5, 20.5
        y1, y2 = min(origin_tf[4], checkfreq[4], gemini[4], variable[4]) - 0.5, \
                 max(origin_tf[4], checkfreq[4], gemini[4], variable[4]) + 0.5
        axins.set_xlim(x1, x2)
        axins.set_ylim(y1, y2)
        
        # Set x-axis to show only "20"
        axins.set_xticks([20])
        axins.set_xticklabels(["20"], fontsize=fontSize)
        axins.set_yticklabels(axins.get_yticks(), fontsize=fontSize)
        
        # Create the small rectangle
        rect = mpatches.Rectangle((19.5, y1), 1, y2 - y1, linewidth=1, edgecolor='r', facecolor='none')
        axs[i].add_patch(rect)
        
        # Manually add connection lines for the large box
        con1 = mpatches.ConnectionPatch(xyA=(1, 1), coordsA=axins.transAxes, 
                                         xyB=(20, origin_tf[4]), coordsB=axs[i].transData, 
                                         color="0.5", linestyle='--')
        axs[i].add_artist(con1)
        
        con2 = mpatches.ConnectionPatch(xyA=(0, 0), coordsA=axins.transAxes, 
                                         xyB=(20, variable[4]), coordsB=axs[i].transData, 
                                         color="0.5", linestyle='--')
        axs[i].add_artist(con2)

    # 获取当前的图例句柄和标签
    handles, labels = axs[0].get_legend_handles_labels()

    # 创建一个黑色的虚拟Line2D对象作为图例项
    gemini_line = Line2D([0], [0], color='grey', marker='x', linestyle='-')
    handles.append(gemini_line)
    labels.append('GEMINI')

    # 更新图例
    fig.legend(handles, labels, fontsize=fontSize * 1.2, loc='upper center', bbox_to_anchor=(0.48, 1.2), ncol=5, frameon=False)
    plt.tight_layout()
    plt.savefig("./exp2.pdf", bbox_inches='tight')
    plt.show()

# 调用函数，并指定 inset_loc 参数
plot_data(inset_loc='upper right')
```
