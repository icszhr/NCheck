```python
import json
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
import matplotlib as mpl

# 确保 Times New Roman 字体可用
font_path = '/usr/share/fonts/truetype/msttcorefonts/Times_New_Roman.ttf'
try:
    fm.fontManager.addfont(font_path)
    mpl.rcParams['font.family'] = 'Times New Roman'
except FileNotFoundError:
    print("Times New Roman font not found, using default serif font。")
    mpl.rcParams['font.family'] = 'serif'
    mpl.rcParams['font.serif'] = 'DejaVu Serif'

# 设置字体和样式
fontSize = 20 / 1.2
x_labels = ['BS 1', 'BS 2', 'BS 4', 'BS 8']
mpl.rcParams.update({
    "font.size": fontSize,
    "axes.labelsize": fontSize * 1.5,
    "xtick.labelsize": fontSize * 1.2,
    "ytick.labelsize": fontSize * 1.2,
    "legend.fontsize": fontSize * 1.2,
    "axes.titlesize": fontSize * 1.5,
    "axes.linewidth": 0.75,
    "xtick.major.width": 0.75,
    "ytick.major.width": 0.75,
    "xtick.major.size": 4,
    "ytick.major.size": 4,
    "lines.linewidth": 1,
    "lines.markersize": 3,
    "figure.figsize": (12, 4),
})

# 设置颜色
colors = ['#B5B5B5', '#D3BFBF', '#8696a7']

# 手动填充数据集
data1 = [
    {'model_name': 'ResNet50', 'total_times': [10, 12, 14, 16]},
    {'model_name': 'ResNet50', 'total_times': [11, 13, 15, 17]},
    {'model_name': 'ResNet50', 'total_times': [9, 11, 13, 15]},
]

data2 = [
    {'model_name': 'ResNet50', 'total_times': [20, 22, 24, 26]},
    {'model_name': 'ResNet50', 'total_times': [21, 23, 25, 27]},
    {'model_name': 'ResNet50', 'total_times': [19, 21, 23, 25]},
]

data3 = [
    {'model_name': 'ResNet50', 'total_times': [30, 32, 34, 36]},
    {'model_name': 'ResNet50', 'total_times': [31, 33, 35, 37]},
    {'model_name': 'ResNet50', 'total_times': [29, 31, 33, 35]},
]

# 提取模型数据
def extract_model_data(data, model_name):
    total_times = []
    for item in data:
        if item['model_name'] == model_name:
            total_times.append(item['total_times'])
    return total_times

# 获取 ResNet50 数据
resnet_data = [
    extract_model_data(data1, 'ResNet50'),
    extract_model_data(data2, 'ResNet50'),
    extract_model_data(data3, 'ResNet50')
]

# 将二维列表展平成一维列表
resnet_data = [times for sublist in resnet_data for times in sublist]

# 绘制图表
def plot_model_data(model_data, model_name):
    bar_width = 0.2
    x = np.arange(len(x_labels))

    fig, ax = plt.subplots(figsize=(8, 6))

    bars1 = ax.bar(x - bar_width, model_data[0], color=colors[0], width=bar_width, edgecolor='grey', label='NCheck')
    bars2 = ax.bar(x, model_data[1], color=colors[1], width=bar_width, edgecolor='grey', label='CheckFreq')
    bars3 = ax.bar(x + bar_width, model_data[2], color=colors[2], width=bar_width, edgecolor='grey', label='GEMINI')

    ax.set_ylabel('Time (s)', fontsize=fontSize * 1.5, weight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(x_labels)
    ax.legend([bars1, bars2, bars3], ['NCheck', 'CheckFreq', 'GEMINI'], fontsize=fontSize * 1.3, loc='upper left', frameon=False)
    ax.tick_params(axis='both', which='major', labelsize=fontSize * 1.5)

    plt.tight_layout()
    plt.savefig('./exp7.pdf', bbox_inches='tight')
    plt.show()

plot_model_data(resnet_data, 'ResNet50')
```
