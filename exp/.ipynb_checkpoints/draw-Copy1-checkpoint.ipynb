```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
import profiler  # profiler.pyPYTHONPATH
```

```python
gpu_fragments, data_transfers = profiler.get_data()
```

```python
# gpu_fragmentsDataFrame
gpu_df = pd.DataFrame(gpu_fragments)

# timestampdatetime
gpu_df['timestamp'] = pd.to_datetime(gpu_df['timestamp'])

# 
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Time')
ax1.set_ylabel('Allocated', color=color)
ax1.plot(gpu_df['timestamp'], gpu_df['allocated'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Fragment', color=color)
ax2.plot(gpu_df['timestamp'], gpu_df['fragment'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
```

```python
# data_transfersDataFrame
transfers_df = pd.DataFrame(data_transfers)

# timestampdatetime
transfers_df['timestamp'] = pd.to_datetime(transfers_df['timestamp'])

# 
gpu2cpu = transfers_df[transfers_df['direction'] == 'GPU2CPU_C']
cpu2gpu = transfers_df[transfers_df['direction'] == 'CPU2GPU_C']

# 
plt.figure(figsize=(10, 6))
plt.plot(gpu2cpu['timestamp'], gpu2cpu['bytes'], label='GPU to CPU')
plt.plot(cpu2gpu['timestamp'], cpu2gpu['bytes'], label='CPU to GPU')

plt.xlabel('Time')
plt.ylabel('Bytes Transferred')
plt.yscale('log')  # Y
plt.legend()
#plt.grid(True, which="both", ls="--")
plt.show()
```

```python
# subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12), sharex=True)

# GPU
color = 'tab:red'
ax1.set_ylabel('Allocated (bytes)', color=color)
ax1.plot(gpu_df['timestamp'], gpu_df['allocated'], color=color, label='Allocated Memory')
ax1.tick_params(axis='y', labelcolor=color)
#ax1.grid(True)

# Y
ax1b = ax1.twinx()
color = 'tab:green'
ax1b.set_ylabel('Fragment', color=color)
ax1b.plot(gpu_df['timestamp'], gpu_df['fragment'], color=color, label='Fragment Ratio')
ax1b.tick_params(axis='y', labelcolor=color)

# GPUCPU
color = 'tab:blue'
ax2.set_ylabel('Bytes Transferred', color=color)
ax2.plot(gpu2cpu['timestamp'], gpu2cpu['bytes'], color='blue', linestyle='-', label='GPU to CPU')
ax2.plot(cpu2gpu['timestamp'], cpu2gpu['bytes'], color='cyan', linestyle='-', label='CPU to GPU')
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_yscale('log')
#ax2.grid(True, which="both", ls="--")

# X
plt.xlabel('Time')

# 
ax1.legend(loc='upper left')
ax1b.legend(loc='upper right')
ax2.legend(loc='upper left')

plt.show()
```
